# Visualization Library Documentation — Matplotlib & Seaborn
## Objective

To create a comprehensive guide on two popular Python data visualization libraries — Matplotlib and Seaborn, focusing on their core features, graph types, use cases, and comparative insights.

## 1. Library Overview
### Matplotlib

Overview:
Matplotlib is a foundational Python visualization library used for creating static, animated, and interactive plots. It provides full control over every visual element, making it ideal for publication-quality graphs.
Key Features:

Low-level, highly customizable.

Integrates well with NumPy & Pandas.

Supports static and 3D visualizations.
Typical Use Cases:

Exploratory data analysis (EDA).

Visual reporting.

Academic and scientific research visuals.

### Seaborn

Overview:
Built on top of Matplotlib, Seaborn simplifies complex visualizations with high-level functions and built-in aesthetic themes.
Key Features:

Cleaner syntax and automatic styling.

Built-in datasets for quick experiments.

Supports statistical visualizations.
Typical Use Cases:

Data exploration and correlation analysis.

Statistical storytelling and data-driven insights.

## 2. Graph Types with Examples
## A. Matplotlib


### 1. Line Plot

Use Case: Tracking trends over time.

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 12, 15, 18, 20]
plt.plot(x, y, color='green', marker='o')
plt.title("Simple Line Plot")
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()
```

### 2. Bar Chart

Use Case: Comparing categorical data.

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]
plt.bar(categories, values, color='orange')
plt.title("Bar Chart Example")
plt.show()
```

### 3. Scatter Plot

Use Case: Identifying relationships between variables.

```python
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color='blue')
plt.title("Scatter Plot Example")
plt.show()
```

### 4. Histogram

Use Case: Showing data distribution.

```python
data = [1,2,2,3,3,3,4,4,4,4]
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram Example")
plt.show()
```

### 5. Pie Chart

Use Case: Displaying proportions in categories.

```python
sizes = [25, 35, 20, 20]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```

## B. Seaborn

### 1. Line Plot

Use Case: Visualizing trends with confidence intervals.

```python
import seaborn as sns
import pandas as pd

data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [3, 4, 6, 8, 10]
})
sns.lineplot(x='x', y='y', data=data)
```

### 2. Scatter Plot

Use Case: Showing variable correlation.

```python
tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex")
```

### 3. Bar Plot

Use Case: Categorical comparisons with built-in stats.

```python
sns.barplot(x="day", y="total_bill", data=tips)
```

### 4. Histogram / Distribution Plot

Use Case: Displaying frequency and density.

```python
sns.histplot(tips["total_bill"], kde=True, color="purple")
```

### 5. Heatmap

Use Case: Visualizing correlations or matrices.

```python
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])
corr = numeric_tips.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

## 3. Comparison: Matplotlib vs Seaborn

| **Feature**                      | **Matplotlib**                   | **Seaborn**                               |
| -------------------------------- | -------------------------------- | ----------------------------------------- |
| **Ease of Use**                  | Moderate (requires more code)    | High (simplified syntax)                  |
| **Customization**                | Excellent (fine-grained control) | Limited compared to Matplotlib            |
| **Aesthetics**                   | Basic styling                    | Built-in attractive themes                |
| **Statistical Support**          | Manual implementation            | Integrated with statistical functions     |
| **Interactivity**                | Basic (via widgets)              | Moderate, better with integration         |
| **Performance (Large Datasets)** | Efficient                        | Slightly slower due to higher abstraction |
